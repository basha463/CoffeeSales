```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#read_csv
df = pd.read_csv('coffee.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
# removing unncessary columns 
df.drop(columns=['Cogs'],axis=1,inplace=True)
df
```

```python
# duplicated columns 
df.duplicated()
```

```python
# statistical summary of all columns 
df.describe()
```

```python
df['Sales'].sum()
```

```python
c=df.groupby('Market')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize= (12,6))
plt.pie(c,labels=c,autopct='%1.1f%%')
plt.title("Market-wise-sales ")
plt.show()
```

```python
df.groupby(['Market','State'])['Sales'].sum().sort_values(ascending=False)
```

```python
d=df.groupby('Type')['Profit'].sum()
plt.figure(figsize= (12,6))
plt.pie(d,labels=d,autopct='%1.1f%%')
plt.title("Type of product and profit")
plt.show()
```

```python
s=df.groupby('Date')['Sales'].sum()
```

```python
plt.figure(figsize=(12,6))
plt.plot(s.index,s.values,marker=0,color='blue')
plt.title('sales trend')
plt.xlabel('Date')
plt.ylabel('Sales')
```

```python
df['Date'] = pd.to_datetime(df['Date'])

s=df.groupby(df['Date'].dt.to_period('M'))['Sales'].sum()
s
```

```python
plt.figure(figsize=(12,6))
plt.plot(s.index.astype(str),s.values,marker='o',color='blue')   # converting dates to string using astype(str). for converting data type use astype()
plt.title('sales trend')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Sales')
```
